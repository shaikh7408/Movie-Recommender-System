# Data preprocessing 

```python
Movies_data['cast'][0]
```

```python
import ast #for converting string list into list
```

```python
Movies_data['genres'][0]
```

```python
#Now filtring only genre od every movie
def genre_filter(genre_list):
    filter_genre=[]
    counter=0
    for i in ast.literal_eval(genre_list):
            #print(i['name'])
            filter_genre.append(i['name'])
    return filter_genre  
```

```python
Movies_data['genres']=Movies_data['genres'].apply(genre_filter)
```

```python
Movies_data['genres']
```

```python
# Now we are filtring top 3 acotrs for every movies
def actor_filter(actor_list):
    filter_actors=[]
    counter=0
    for i in ast.literal_eval(actor_list):
        if counter!=3:
            #print(i['name'])
            filter_actors.append(i['name'])
            counter=counter+1
        else:
            break;
    return filter_actors   
```

```python
Movies_data['cast']=Movies_data['cast'].apply(actor_filter)
```

```python
#top 3 acotr for every movie
Movies_data['cast']
```

```python
Movies_data
```

```python
Movies_data['crew'][0]
```

```python
# In crew feature or column we want only  director name nothling else we have filter director name
def director_filter(director_list):
    filter_director=[]
    for i in ast.literal_eval(director_list):
        if i['job']=='Director':
            #print(i['name'])
            filter_director.append(i['name'])
    return filter_director  
```

```python
Movies_data['crew']=Movies_data['crew'].apply(director_filter)
```

```python
Movies_data['crew']
```

```python
Movies_data.head(3)
```

```python
Movies_data['keywords'][0]
```

```python
# Now we have to filter only keyword for movies
def keyword_filter(keyword_list):
    filter_keyword=[]
    for i in ast.literal_eval(keyword_list):
            #print(i['name'])
            filter_keyword.append(i['name'])
    return filter_keyword
```

```python
Movies_data['keywords']=Movies_data['keywords'].apply(keyword_filter)
```

```python
Movies_data['keywords'][0]
```

```python
Movies_data.head(2)
```

```python
#now removing the space in between word because we want every word uniqe meaning
Movies_data['genres']=Movies_data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
Movies_data['cast']=Movies_data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
Movies_data['keywords']=Movies_data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
Movies_data['crew']=Movies_data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python

Movies_data['overview'][0]
```

```python
#Making Overview string into list
Movies_data['overview']=Movies_data['overview'].apply(lambda x:str(x).split())
```

```python
Movies_data.head(2)
```

```python
#Now create tag for every movie becuase We want only one tag for every movie
```

```python
Movies_data['tag']=Movies_data['overview']+Movies_data['genres']+Movies_data['keywords']+Movies_data['cast']+Movies_data['crew']
```

```python
Movies_data.head(2)
```

```python
Movies_data['tag'][0]
```

```python
Movies_data_new=Movies_data[['id','title','tag']]
```

```python
Movies_data_new.head()
```

```python
# now making tag list into string
Movies_data_new['tag']=Movies_data_new['tag'].apply(lambda x:" ".join(x))
```

```python
#convert every tage into lowercase
Movies_data_new['tag']=Movies_data_new['tag'].apply(lambda x:x.lower())
```

```python
Movies_data_new['tag'][0]
```

```python
Movies_data_new.head()
```

```python
# now  we have to do Text vectroization for model building
```

```python
import nltk
```

```python
# frist we have to perform stamming in tag column
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
# frist we have to perform stamming in tag column
def stamming(tag):
    stamm=[]
    for i in tag.split():
        stamm.append(ps.stem(i))
    return " ".join(stamm)                    
```

```python
Movies_data_new['tag']=Movies_data_new['tag'].apply(stamming)
```

```python
Movies_data_new.head()
```

```python
# now  we have to do Text vectroization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector_tag=CV.fit_transform(Movies_data_new['tag']).toarray()
```

```python
vector_tag
```

```python
vector_tag.shape
```

```python
# Now we have caluclate cosing similrity between every movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Now we have caluclate cosing similrity between every movies
similarity_score=cosine_similarity(vector_tag)
```

```python
# create Recommended moive main function 
def movie_recommend(movie):
    moive_index=Movies_data_new[Movies_data_new['title']==movie].index[0]
    Recommended_moive=sorted(enumerate(similarity_score[moive_index]),reverse=True,key=lambda x:x[1])[0:10]
    print(Recommended_moive)
    for i in Recommended_moive:
        #print(Movies_data_new.iloc[i[0]].title)
        print(Movies_data_new['title'][i[0]])
```
